```python
# Import ords_prods_merged.pkl

df_ords_prods = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))
```

```python
# Change data type

df_ords_prods['product_id']=df_ords_prods['product_id'].astype('float16')
df_ords_prods['add_to_cart_order']=df_ords_prods['add_to_cart_order'].astype('float16')
df_ords_prods['reordered'] =df_ords_prods['reordered'].astype('float16')
df_ords_prods['aisle_id']=df_ords_prods['aisle_id'].astype('int8')
df_ords_prods['department_id']=df_ords_prods['department_id'].astype('int8')
df_ords_prods['prices']=df_ords_prods['prices'].astype('float16')
```

# 1.1 Create the 'price_label' column

```python
df_ords_prods.loc[df_ords_prods['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods.loc[(df_ords_prods['prices'] <= 15) & (df_ords_prods['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df_ords_prods.loc[df_ords_prods['prices'] <=5, 'price_range_loc'] = 'Low-range product'
```

```python
#Display the product ranges

df_ords_prods['price_range_loc'].value_counts(dropna = False)
```

# 1.2 Create the 'busiest_day' column

```python
# Get the frequency of the values in the 'orders_day_of_week' column

df_ords_prods['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create a for-loop that append to the 'result' list the busyness of the days based on values in the 'orders_day_of_week' column
result = []

for value in df_ords_prods["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Create new column ' busiest_day in df_ords_prods'
df_ords_prods['busiest_day'] = result
```

# 2. Create the 'busiest_days columns'

```python
# Get the frequency of the values in the 'orders_day_of_week' column
df_ords_prods['orders_day_of_week'].value_counts(dropna = False)
```

### Observation
The two busiest day are: 0 ( Saturday) and 1 ( Sunday)
The two slowest day are: 4( Wednesday) and 3 (Tuesday)

```python
# Create a for-loop that append to the 'days' list the busyness of the days based on values in the 'orders_day_of_week' column

days = []

for day in df_ords_prods['orders_day_of_week']:
  if day == 0 or day == 1:
    days.append('Busiest days')
  elif day == 4 or day == 3:
    days.append('Slowest days')
  else:
    days.append('Regular days')
```

```python
# Create new column ' busiest_days in df_ords_prods'
df_ords_prods['busiest_days'] = days
```

```python
# Get the frequency of the values in the 'busiest_days' column
df_ords_prods['busiest_days'].value_counts(dropna = False)
```

```python
# 3.  Check the output of the 'orders_day_of_week' and the 'busiest_days' column for accuracy

print('Output of the orders_day_of_week and the busiest_days column:\n', df_ords_prods[['orders_day_of_week', 'busiest_days']].head(10))
```

# 4. Create the 'busiest_period_of_day' column


```python
# Get the frequency of the values in the 'order_hour_of_day' column
df_ords_prods['order_hour_of_day'].value_counts(dropna = False)
```

### Observation:
Most Orders in hours: 10,11,14,15,13,12,16,9;
Average Orders in hours: 17,8,18,19,20,7,21,22;
Fewest Orders in hours: 23,6,0,1,5,2,4,3.

```python
# create a for-loop that append to the 'hours' list the busyness of the hours based on values in the 'order_hour_of_day' column

hours = []

for hour in df_ords_prods["order_hour_of_day"]:
  if hour in [10,11,14,15,13,12,16,9]:
    hours.append("Most orders")
  elif hour in [17,8,18,19,20,7,21,22]:
    hours.append("Average orders")
  else:
    hours.append("Fewest orders")
```

```python
# Create new column ' busiest_period_of_day ' in df_ords_prods
df_ords_prods['busiest_period_of_day'] = hours
```

```python
# Check the output of the 'order_hour_of_day' and the 'busiest_period_of_day' column for accuracy
df_ords_prods[['order_hour_of_day', 'busiest_period_of_day']].head(10)
```

```python
# 5. Get the frequency of the values in the 'busiest_period_of_day' column

df_ords_prods['busiest_period_of_day'].value_counts(dropna = False)
```
